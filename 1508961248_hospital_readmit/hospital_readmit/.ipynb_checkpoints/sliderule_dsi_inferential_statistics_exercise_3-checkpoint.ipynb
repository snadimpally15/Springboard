# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

```python
# Null hypothesis: r = 0.
hyp_r = 0
n = len(df.Discharges)

# Observed corr coeff
obs_r, obs_p_value = stats.pearsonr(df.Discharges, df['Excess Readmissions'])

bs_r = np.empty(10000) 
bs_p = np.empty(10000)

# generate pairs bootstrap samples
for i in range(10000):
   idx = np.random.choice(range(n), n)
   bs_r[i], bs_p[i] = stats.pearsonr(df.Discharges, df['Excess Readmissions'])

bs_r_mean = np.mean(bs_r)
bs_p_mean = np.mean(bs_p)

print('BOOTSTRAP RESULTS\nBootstrap r: \t \t {:0.5}\n|obs r - bs r| : \t {:0.5}'.format(bs_r_mean, np.abs(obs_r - bs_r_mean)))

print('\nBootstrap p-value: \t {:0.5}'.format(bs_p_mean))
```

## Statistical Significance

Compute and report the observed significance value (or p-value) and report statistical significance for 
α = .01.

                        t = r*np.sqrt(n - 2) / np.sqrt(1 - r**2)


```python
# calculate the correlation coefficient
r = np.corrcoef(df.Discharges, df['Excess Readmissions'])[0][1]

# since r = 0 the formula for test statistic becomes: r * √(n-2)/(1-r**2)
n = len(df.Discharges)
t_stat = r * np.sqrt(n-2)/np.sqrt(1-r**2)
p_value = stats.t.cdf(x=t_stat, df = n-2) * 2

print('observed r: \t {:0.5}\n\nFREQUENTIST RESULTS\nt-statistic: \t {:0.5}\np-value: \t {:0.5}'.format(r, t_stat, p_value))
```

The p-value under both the frequentist and bootstrap tests is extremely low, so the null hypothesis that there is no correlation between hospital size and readmission rate must be rejected. The correlation coefficient is -0.093, so there is a slight negative correlation between the size of a hospital and its excess readmissions; readmission rates drop slightly as the number of discharges increases. The threshold for the readmission rate to be deemed excessive is 1. The mean of the smallest hospitals' readmission rate is less than 2% above the threshold, and less than 4% above the mean readmission rate of the large hospitals.

## Statistical v Practical Significance

Discuss statistical significance and practical significance. Do they differ here? How does this change your recommendation to the client?

While the analysis appears to be correct, I disagree with the recommendation to require consolidation of smaller facilities into larger facilities. Further considerations need to be determined, but it looks as though the drop in the readmission rate, while real and statistically significant, may not have a sufficient practical significance to warrant the recommendations to upgrade smaller facilities and consolidate into larger facilities. In other words, this might be an instance where statistical significance does not mean practical significance; the proposed fix might be orders of magnitude more costly than continuing on the current course.
